```python
import numpy as np
import string
import random
```

```python
prob = np.zeros([27, 27, 27, 27]) # third letter, second letter, first letter, probability of next letter where 0 = END  and 1-26 = a-z


#data_file =  open(r'C:/Users/matcb/Downloads/words.txt')
data_file =  open(r'C:/Users/matcb/Downloads/drugs.txt')

for line in data_file:
    
    third = 0
    second = 0
    first = 0
    
    clean_line = line.strip().lower()

    if clean_line.isalpha():
        
        for letter in clean_line:

            prob[ third, second, first, ord(letter)-96 ] = prob[ third, second, first, ord(letter)-96 ] + 1

            third = second
            second = first
            first = ord(letter)-96
            
        prob[third, second, first, 0 ] =  prob[third, second, first, 0] + 1 #end of the word
        
data_file.close()            
```

```python
n=0
while n < 50:
    new_word = ''

    last = 0
    last_two = 0


    first = random.choices(' abcdefghijklmnopqrstuvwxyz', weights=prob[last_two][last][0])[0]
    new_word = new_word + first


    i = 0
    while i < 15:
        new_letter = random.choices(' abcdefghijklmnopqrstuvwxyz', weights=prob[last_two][last][ord(first)-96])[0]
        if new_letter == " ":
            break
        new_word = new_word + new_letter
        last_two = last
        last = ord(first)-96
        first  = new_letter

        i = i+1

    data_file =  open(r'C:/Users/matcb/Downloads/drugs.txt')
    wordlist = data_file.read().lower()
    if new_word not in wordlist:
        print (new_word)
        n=n+1
        
```
